# StandardScaler 

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler


X = np.array([[1, 5],
              [2, 8],
              [3, 6],
              [4, 9],
              [5, 10]])

# Standaryzacja danych
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Wykres danych przed i po standaryzacji
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], color='blue')
plt.title("Przed standaryzacjÄ…")
plt.xlabel("Cecha A")
plt.ylabel("Cecha B")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], color='green')
plt.title("Po standaryzacji")
plt.xlabel("Cecha A (standaryzowana)")
plt.ylabel("Cecha B (standaryzowana)")
plt.grid(True)

plt.tight_layout()
plt.show()
```
